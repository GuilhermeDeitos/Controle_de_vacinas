## Controle de carteiras de vacinação

Desenvolva um sistema que consiste no registro e busca de informações em carteiras de vacina de pessoas. O sistema deve possuir um menu com cadastro de pessoa, registro de vacina, busca e relatório.

O sistema deve possibilitar:
- cadastro de pessoas, com:
    - nome (*string*)
    - RG (*string*)
- referente a cada pessoa, deve possuir o registro de vacinas específicas, com registro de:
    - doença que a vacina protege (*string*)
    - data de aplicação da vacina (*string* dd/mm/aaaa)
    - dose tomada (inteiro)
- busca por:
    - nome da pessoa
    - RG
- relatórios com:
    - total de pessoas vacinadas para cada doença
    - total de pessoas vacinadas por mês
    - média de vacinas por pessoa


O sistema deve possibilitar salvar dados cadastrados em arquivo, e recuperar o arquivo durante a abertura do sistema. 

No desenvolvimento, arquivos, listas e dicionários devem ser utilizados. Os arquivos podem ser de qualquer tipo (texto, csv, json ou outros)

Deve ser entregue:
- O projeto (Jupyter ou Colaboratory);
- Um arquivo, com dados a serem carregados quando a execução do sistema for iniciada.

```python
import csv
```

```python
#Cores para embelezar

black =	'\033[1;30m'

red =	'\033[1;31m'

green	= '\033[1;32m'

yellow	= '\033[1;33m'

blue	= '\033[1;34m'

purple	= '\033[1;35m'

cyan	= '\033[1;36m'

reset =	'\033[0;0m'
```

```python
def cadastroDePessoas(pessoa,aux):
  with open('Pessoas.csv','a') as arq: #abrir o arquivo csv
    arq.writelines('Nome,RG,Vacina,Dia,Mês,Ano,Dose\n') #apenas para deixar de cabeçario 
    lista = ['Nome','RG','Vacina','Dia','Mês','Ano','Dose'] #a lista de campos 
    writer = csv.DictWriter(arq,fieldnames=lista) #A função para escrever no arquivo csv usando dicionarios
    cont = int(input('Quantas pessoas deseja cadastrar?')) #contador para ver quantas pessoas deseja cadastrar
    for i in range(1, cont + 1): #loop
      #pedindo as informações
      pessoa['Nome'] = input(f'Informe o nome da {i}ª pessoa ')
      pessoa['RG'] = input(f'Informe o RG da {i}ª pessoa ')
      pessoa['Vacina'] = input('Informe qual doença a vacina protege ').upper()
      pessoa['Dia'] = input('Informe o dia da aplicação da vacina ')
      pessoa['Mês'] = input('Informe o mês da aplicação da vacina ')
      pessoa['Ano'] = input('Informe o dia da aplicação da vacina ')
      pessoa['Dose'] = int(input('Informe qual a Dose tomada na vacina '))
      writer.writerow(pessoa) #passando as informações para o arquivo csv
      aux +=1 #um contador para saber quantos foram vacinados no final

      
```

```python
def buscar(pessoa):
  print(purple,'=-=-=-=-=-=- Menu Busca =-=-=-=-=-=-\n'
  '[1]Busca por Nome\n'
  '[2]Busca por RG\n'
  '[0]Sair',reset) #Menu bonitinho
  op2 = int(input())
  while True:
    if op2 == 1:
      busca = input('Informe o nome para busca: ') #caso deseje buscar nome
      return search(pessoa,busca) #retornar a função que fiz para busca
      break
    elif op2 == 2:
      busca = input('Informe o RG para busca: ')#caso deseje buscar RG
      return search(pessoa,busca)#retornar a função que fiz para busca
      break
    elif op2 == 0:
      return print(blue,'SAINDO...')
      break
    else: #Caso não seja nenhum dos valores acima é invalido e fecha essa tela
      return print(red,'OPÇÃO INVALIDA')
      break  
    

    
```

```python
def search(pessoa,busca): #função de busca 
  with open('Pessoas.csv','r') as arq: #abrindo o arquivo para leitura
    lista = ['Nome','RG'] #só quero os campos nome e rg então fiz uma lista só com eles
    reader = csv.DictReader(arq,fieldnames=lista)

    for aux in reader: #um loop para adentrar o dict
      for k,v in aux.items():
        if busca == v: #caso o valor sejá igual ele irá retornar encontrado e caso não ele não retornará nada
          return 'Encontrado',aux.items()
          break


```

```python
def MostrarPessoas(pessoa):
  #apenas para eu ter uma ideia das informações, não foi pedido no exercicio
  with open('Pessoas.csv','r') as arq:
    lista = ['Nome','RG','Vacina','Dia','Mês','Ano','Dose']
    reader = csv.DictReader(arq,fieldnames=lista)

    for r in reader:
      for k,i in r.items():
        print(f'{green}{k}: {i}') #Mostrará todos os itens da tabela
```

```python
def relatorio(pessoa,qtdData,qtdVacinas):
  print(purple,'=-=-=-=-=-=- Menu Relatório =-=-=-=-=-=-\n'
  '[1]Relatório de total de pessoas vacinadas para cada doença\n'
  '[2]Relatório de total de pessoas vacinadas por mês \n'
  '[0]Sair',reset) #Menu bonitinho
  op2 = int(input())
  while True:
    if op2 == 1:
      aux = relatorioVacinas(pessoa,qtdVacinas) #retornar a função que fiz para o relatório das vacinas
      for k,v in aux.items(): #for para deixar mais bonito e organizado
        print(f'{blue} {k}: {v}')
      break

    elif op2 == 2:
      aux = relatorioData(pessoa,qtdData) #retornar a função que fiz para o relatório dos meses e quantidade de vacinas
      for k,v in aux.items(): #for para deixar mais bonito e organizado
        print(f'{blue} {k}: {v}')
      break
    elif op2 == 0:
      return print(blue,'SAINDO...')
      break
    else: #Caso não seja nenhum dos valores acima é invalido e fecha essa tela
      return print(red,'OPÇÃO INVALIDA')
      break  
```

```python
def relatorioData(pessoa,qtdData):

  qtdData = {
        'Janeiro':0,
        'Fevereiro':0,
        'Março':0,
        'Abril':0,
        'Maio':0,
        'Junho':0,
        'Julho':0,
        'Agosto':0,
        'Setembro':0,
        'Outubro':0,
        'Novembro':0,
        'Dezembro':0
      }

  with open('Pessoas.csv','r') as arq:#abrir o arquivo
    lista = ['Nome','RG','Vacina','Dia','Mês','Ano','Dose']
    reader = csv.DictReader(arq,fieldnames=lista)
    
      
    
    for j in reader:#Procurar no arquivo os Meses e verificar se é correspondente, no que for ele add +1 no contador
      if j['Mês'] == '01':
        qtdData['Janeiro'] += 1
      if j['Mês'] == '02':
        qtdData['Fevereiro'] += 1
      if j['Mês'] == '03':
        qtdData['Março'] += 1
      if j['Mês'] == '04':
        qtdData['Abril'] += 1
      if j['Mês'] == '05':
        qtdData['Maio'] += 1
      if j['Mês'] == '06':
        qtdData['Junho'] += 1
      if j['Mês'] == '07':
        qtdData['Julho'] += 1
      if j['Mês'] == '08':
        qtdData['Agosto'] += 1
      if j['Mês'] == '09':
        qtdData['Setembro'] += 1
      if j['Mês'] == '10':
        qtdData['Outubro'] += 1
      if j['Mês'] == '11':
        qtdData['Novembro'] += 1
      if j['Mês'] == '12':
        qtdData['Dezembro'] += 1

      
  return qtdData     #Retornas o Dict com as informações da quantidade de vacinas por mês
```

```python
def relatorioVacinas(pessoa,qtdVacinas):
  qtdVacinas = {
    'Hepatite B':0,
    'Febre Amarela':0,
    'Tétano':0,
    'HPV':0,
    'Gripe':0
  }

  with open('Pessoas.csv','r') as arq: #abrir o arquivo
    lista = ['Nome','RG','Vacina','Dia','Mês','Ano','Dose']
    reader = csv.DictReader(arq,fieldnames=lista)

    for i in reader: #Procurar no arquivo as doenças e verificar se a vacina é correspondente, no que for ele add +1 no contador
      if  i['Vacina'] == 'GRIPE':
        qtdVacinas['Gripe'] += 1
      if  i['Vacina'] == 'HPV':
          qtdVacinas['HPV'] += 1
      if i['Vacina'] == 'TETANO':
          qtdVacinas['Tétano'] += 1
      if i['Vacina'] == 'FEBRE AMARELA':
          qtdVacinas['Febre Amarela'] += 1
      if i['Vacina'] == 'HEPATITE B':
          qtdVacinas['Hepatite B'] += 1

  return qtdVacinas #Retornas o Dict com as informações da quantidade de vacina
```
